```python
import pandas as pd
import numpy as np
```

```python
raw = pd.read_csv("dataset 2.csv")
raw.shape
raw.info()
```

```python
columns = [ "track_id", 
           "popularity", 
           "duration_ms", 
           "danceability", 
           "energy", 
           "key", 
           "loudness", 
           "speechiness", 
           "liveness", 
           "valence", 
           "time_signature", 
           "track_genre"]

df = raw[columns].copy()
```

```python
df = df.drop_duplicates(subset = "track_id")
df = df.dropna()
```

```python
df["track_genre"] = (
    df["track_genre"]
    .astype(str)
    .str.lower()
    .str.strip()
)
```

```python
allowed_genres = [
    "acoustic", "classical", "dance", "disco",
    "happy", "hiphop", "indian", "jazz",
    "pop", "romance", "sad"
]

df = df[df["track_genre"].isin(allowed_genres)].copy()
```

```python
df["track_genre"].value_counts()
```

```python
audio_features = ["danceability", "energy", "valence"]

df[audio_features].describe()
```

```python

df["is_happy_genre"] = df["track_genre"] == "happy"
df["is_sad_genre"] = df["track_genre"] == "sad"
```

```python
# High valence = emotionally positive
df["high_valence"] = df["valence"] >= 0.6

# Low valence = emotionally negative
df["low_valence"] = df["valence"] <= 0.4
```

```python
df.info()
df.head()
```

```python
df.to_csv("tracks_cleaned.csv", index=False)
```
