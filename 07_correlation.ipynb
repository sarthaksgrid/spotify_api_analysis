```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('tracks_cleaned.csv')

metrics = ['popularity', 'danceability', 'energy', 'loudness', 'speechiness', 'liveness', 'valence']
available_metrics = [m for m in metrics if m in df.columns]

genre_metrics = df.groupby('track_genre')[available_metrics].mean()

scaled_df = (genre_metrics - genre_metrics.min()) / (genre_metrics.max() - genre_metrics.min())

plt.figure(figsize=(12, 25)) 
sns.heatmap(
    scaled_df, 
    annot=genre_metrics, 
    fmt=".2f",           
    cmap='YlGnBu',       
    square=True,        
    cbar=True,
    linewidths=.5,
    annot_kws={"size": 8} 
)

plt.title('Musical DNA: Genre Metrics (Square Grid)', fontsize=16, pad=20)
plt.show()
```
